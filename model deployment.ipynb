```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
```

```python
model = tf.keras.models.load_model("mobilenet_V2modelon3ds V0.2.h5")
```

```python
class_names = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
def preproses(image):
    image = tf.cast(image, tf.float32) / 255.0
    return image
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        # Extract and preprocess the face
        face = frame[y:y+h, x:x+w]
        face_resized = cv2.resize(face, (224, 224))
        face_rgb = cv2.cvtColor(face_resized, cv2.COLOR_BGR2RGB)
        face_preprocessed = preproses(face_rgb)
        face_batch = np.expand_dims(face_preprocessed, axis=0)

        # Predict
        predictions = model.predict(face_batch)
        predicted_class = class_names[np.argmax(predictions)]

        # Draw result
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, ), 2)
        cv2.putText(frame, predicted_class, (x, y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Show webcam
    cv2.imshow("Real-Time FER - MobileNetV3", frame)

    # Exit with 'q'
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
